# Data

```python
np_train = np.load('Titanic_data_train.npz')
train_inputs = np_train['inputs'].astype(np.float)
tain_targets = np_train['targets'].astype(np.int)
```

```python
np_val = np.load('Titanic_data_val.npz')
val_inputs = np_val['inputs'].astype(np.float)
val_targets = np_val['targets'].astype(np.int)
```

```python
np_test = np.load('Titanic_data_test.npz')
test_inputs = np_test['inputs'].astype(np.float)
```

# Model

```python
input_size = 9
output_size = 1
hidden_layer_size = 100

model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='sigmoid')
])

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


batch_size = 100
max_epochs = 75

model.fit(train_inputs,
         train_targets,
         batch_size = batch_size,
         epochs = max_epochs,
         validation_data = (val_inputs, val_targets),
         verbose = 2)
```

```python
test_loss, test_accuracy = model.evaluate(test_inputs)
```

```python
test_preds = model.predict(test_inputs)
```
